```python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
```

```python
import matplotlib.pyplot as plt

# Remove duplicate 'fips' column
df = df.loc[:,~df.columns.duplicated()]

# Create a pivot table with 'year' as index, 'name' as column and 'income' as values
pivot_df = df.pivot(index='year', columns='name', values='income')

# Calculate the growth rate 
growth_rate_df = pivot_df.pct_change()

# Plot the growth rate for all states
plt.figure(figsize=(15,10))
for column in growth_rate_df:
    plt.plot(growth_rate_df.index, growth_rate_df[column], marker='', linewidth=1, alpha=0.9, label=column)

plt.legend(loc=2, ncol=2)
plt.title("Income growth rate for all states", loc='left', fontsize=12, fontweight=0, color='orange')
plt.xlabel("Year")
plt.ylabel("Growth Rate")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Remove duplicate 'fips' column
df = df.loc[:,~df.columns.duplicated()]

# Create a pivot table with 'year' as index, 'name' as column and 'income' as values
pivot_df = df.pivot(index='year', columns='name', values='income')

# Calculate the growth rate 
growth_rate_df = pivot_df.pct_change()

# Plot the growth rate for all states
plt.figure(figsize=(15,10))
for column in growth_rate_df:
    plt.plot(growth_rate_df.index, growth_rate_df[column], marker='', linewidth=1, alpha=0.9, label=column)

plt.legend(loc=2, ncol=2)
plt.title("Income growth rate for all states", loc='left', fontsize=12, fontweight=0, color='orange')
plt.xlabel("Year")
plt.ylabel("Growth Rate")
plt.show()
```

```python
import matplotlib.pyplot as plt

# calculate income growth rate by state
df['income_growth'] = df.groupby('name')['income'].pct_change() * 100

# plot income growth rate
plt.figure(figsize=(15,8))
for state in df['name'].unique():
    state_data = df[df['name'] == state]
    plt.plot(state_data['year'], state_data['income_growth'], label=state)

plt.xlabel('Year')
plt.ylabel('Income Growth Rate (%)')
plt.title('Income Growth Rate by State from 2005 to 2019')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
